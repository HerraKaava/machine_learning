```python
import numpy as np
```

```python
X = np.random.normal(size=(5,3))
rgen = np.random.RandomState(1)
w = rgen.normal(loc=0.0, scale=0.01, size=X.shape[1])
b = np.float64(0.0)
y = np.random.choice(10, size=5)
```

```python
print(X.shape)
print(w.shape)
print(b.shape)
print(y.shape)
```

```python
output = X @ w + b
errors = y - output
```

### Loop for updating the weights in gradient descent (inefficient)

```python
w
```

```python
n = X.shape[0]
p = X.shape[1]
eta = 0.01
for j in range(p):
    for i in range(n):
        w[j] += eta * 2.0 * errors[i] * X[i, j] / n
```

```python
# Updated weights after one iteration of full batch gradient descent
w
```

### Vectorized version for updating the weights in gradient descent (efficient)

```python
# Initialize the same starting weights as above
rgen = np.random.RandomState(1)
w = rgen.normal(loc=0.0, scale=0.01, size=X.shape[1])
```

```python
w
```

```python
w += eta * 2.0 * X.T @ errors / n
```

```python
# Updated weights after one iteration of full batch gradient descent
w
```

- As can be seen, the result is the same.
